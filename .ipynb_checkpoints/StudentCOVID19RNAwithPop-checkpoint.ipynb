# Introduction to Programming

```python
import random
import numpy as np
import sys
import matplotlib.pyplot as plt
from datetime import datetime  
from datetime import timedelta  
```

```python
#mutation rate in hours
mutrate = 1/200
mutcount=0
totalpop=1
nucDict=["A", "T", "G", "C"]
seqlen=20
datestart=datetime(2020,2,1)
growthrate=1.2
totdays=90
```

```python
#Define Tree Class
class RNA(object):
    def __init__(self, name, pop, frac, loc, time, parent, country=None):
        self.name = "Strain_" + str(name)
        self.NameInt = name
        self.pop  = pop
        self.frac = frac
        self.loc = loc
        self.time = time
        self.children = []
        self.parent=parent
        self.rnaint= [0 for i in range(0,seqlen)]
        self.rna =""
        if country == None:
            self.country="panama"

    def add_child(self, obj):
        self.children.append(obj)
    def rna_string(self):
        self.rna =""
        for e in self.rnaint:
            self.rna= self.rna+nucDict[e]
    def headerandseq(self,file):
        print(">"+self.name,file=file)
        print(self.rna,file=file)
    def seqdat(self,file):
        newdate=datestart+timedelta(self.time)
        print(self.name+"\t"+"COVID19"+"\t"+str(newdate.date())+"\t"+"panama",file=file)
    def date(self):
        newdate=datestart+timedelta(self.time)
        return str(newdate.date())
    def mutate(self):
        self.rnaint=self.parent.rnaint.copy()
        self.rnaint[self.loc]= (self.rnaint[self.loc]+random.randrange(3)+1)%4
        self.rna_string()
```

```python
#Tree printer
def pprint_tree(node, file=None, _prefix="", _last=True):
    print(_prefix, "|--" if _last else "|--", "S"+str(node.NameInt)+" T: " +node.date() +"P: "+ str(int(node.pop)), sep="", file=file)
    _prefix += "        " if _last else "|        "
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        pprint_tree(child, file, _prefix, _last)
        
        
def pprint_rna_tree(node, file=None, _prefix="", _last=True):
    print(_prefix, "|--" if _last else "|--", node.rna, sep="", file=file)
    _prefix += "        " if _last else "|        "
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        pprint_rna_tree(child, file, _prefix, _last)
        
#NextStrainPrint        
def nextstrain_print(node, file=None, _last=True):
    node.headerandseq(file)
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        nextstrain_print(child, file, _last)   
        
# Strain Virus Date  
def nextstrain_print_dat(node, file=None, _last=True):
    node.seqdat(file)
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        nextstrain_print_dat(child, file, _last)

        
#Genome mutate
def mutate_tree(node, file=None, _last=True):
    node.rnaint=node.parent.rnaint.copy()
    child_count = len(node.children)
    
    thisnode=node.loc
    thisgene=node.rnaint[thisnode]
    node.rnaint[thisnode]= (node.rnaint[thisnode]+random.randrange(3)+1)%4
    node.rna_string()
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        mutate_tree(child, file, _last)
        
```

```python
# Generate Starting sequence: First column holds count
root=RNA("root", 0, 0 ,0, 0, 0)
for i in range(0,seqlen):
    root.rnaint[i]= random.randrange(4)
root.rna_string()
    
initRNA = RNA(0, 1, 1, 0, 0, root)
initRNA.mutate()

    
root.add_child(initRNA)
genomelist = [initRNA]

#Simulate Population Growth
for t in range(0,totdays):
    mutates= random.random()
    
    if mutates <totalpop*(growthrate-1)*mutrate:
        mutcount=mutcount+1
        prob=[]
        for rna in genomelist:
            prob.append(rna.frac)
            
            
        whichparent=random.choices(genomelist, prob)[0]
        newsequence=RNA(mutcount,1,0,random.randrange(0,seqlen),t, whichparent)
        newsequence.mutate()
        genomelist.append(newsequence)
        whichparent.add_child(newsequence)
        totalpop=totalpop+1
        
    totalpop=growthrate*totalpop
    for rna in genomelist:
        rna.pop=rna.pop*growthrate
        rna.frac=rna.pop/totalpop
        
      
        

print("mutations: ",mutcount, "\ntotal infected:", int(totalpop))
#mutate_tree(initRNA)
pprint_tree(initRNA)
pprint_rna_tree(initRNA)
fasta = open('data//sequences.fasta', 'w') 
nextstrain_print(initRNA,file=fasta)
fasta.close()

meta= open('data//metadata.tsv','w')
print("strain"+"\t"+"virus"+"\t"+"date"+"\t"+"country",file=meta)
nextstrain_print_dat(initRNA,file=meta)
meta.close()

```

```python
def pprint_tree_sampled(node, file=None, _prefix="", _last=True):
    if node.pop > 0.01*random.random()*totalpop:
        print(_prefix, "|--" if _last else "|--", "S"+str(node.NameInt)+" T: " +node.date() +"P: "+ str(int(node.pop)), sep="", file=file)
        _prefix += "        " if _last else "|        "
        child_count = len(node.children)
        for i, child in enumerate(node.children):
            _last = i == (child_count - 1)
            pprint_tree_sampled(child, file, _prefix, _last)
print(totalpop*0.05)
pprint_tree_sampled(initRNA)
```
