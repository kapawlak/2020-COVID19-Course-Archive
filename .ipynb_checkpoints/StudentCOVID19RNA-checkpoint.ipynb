````python
import random
import numpy as np
import sys
import matplotlib.pyplot as plt
from datetime import datetime  
from datetime import timedelta
import BasicStochasticSIR as SIR
````

````python
#Define Tree Class
class RNA(object):
    def __init__(self, index=None, pop=0, nuc=None, time=None, parent=None, country= None):
        self.index = index
        self.pop  = pop
        self.Pop=[]
        self.nuc = nuc
        self.time = time
        self.children = []
        self.parent=parent
        self.rnaint= [0 for i in range(0,seqlen)]
        self.rna =""
        if country == None:
            self.country="panama"

    def add_child(self, obj):
        self.children.append(obj)
        obj.parent = self
        
    def name(self):
        return "Strain_" + str(self.index)
    
    def populate(self, days, t):
        #S, I, R = SIR.quick_infected(maxdays=days+1)
        #print(I[0])
        self.Pop= [0 for i in range(0,t)] + [(1/5)*(i**4)*(2**(i/10 - (i/10)*(i/10))) for i in range(0,days-t+1)]
        #print(self.Pop)
        
    def rna_string(self):
        self.rna =""
        for e in self.rnaint:
            self.rna= self.rna+["A", "T", "G", "C"][e]
    def headerandseq(self,file):
        print(">"+self.name(),file=file)
        print(self.rna,file=file)
    def seqdat(self,file):
        newdate=datestart+timedelta(self.time)
        print(self.name()+"\t"+"COVID19"+"\t"+self.date()+"\t"+"panama",file=file)
    def date(self):
        newdate=datestart+timedelta(self.time)
        return str(newdate.date())
    def mutate(self):
        self.rnaint=self.parent.rnaint.copy()
        self.rnaint[self.nuc]= (self.rnaint[self.nuc]+random.randrange(3)+1)%4
        self.rna_string()
````

````python
#Tree printer
def pprint_tree(node, file=None, _prefix="", _last=True):
    print(_prefix, "|--" if _last else "|--", "S"+str(node.index)+" T: " +node.date() +"P: "+ str(int(node.pop)), sep="", file=file)
    _prefix += "        " if _last else "|        "
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        pprint_tree(child, file, _prefix, _last)
        
        
def pprint_rna_tree(node, file=None, _prefix="", _last=True):
    print(_prefix, "|--" if _last else "|--", node.rna, sep="", file=file)
    _prefix += "        " if _last else "|        "
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        pprint_rna_tree(child, file, _prefix, _last)
        
#NextStrainPrint        
def nextstrain_print(node, file=None, _last=True):
    node.headerandseq(file)
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        nextstrain_print(child, file, _last)   
        
# Strain Virus Date  
def nextstrain_print_dat(node, file=None, _last=True):
    node.seqdat(file)
    child_count = len(node.children)
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        nextstrain_print_dat(child, file, _last)

        
#Genome mutate
def mutate_tree(node, file=None, _last=True):
    node.rnaint=node.parent.rnaint.copy()
    child_count = len(node.children)
    
    thisnode=node.nuc
    thisgene=node.rnaint[thisnode]
    node.rnaint[thisnode]= (node.rnaint[thisnode]+random.randrange(3)+1)%4
    node.rna_string()
    for i, child in enumerate(node.children):
        _last = i == (child_count - 1)
        mutate_tree(child, file, _last)
#outgroup        
def outgroup(node, file=None):
     print(">Outgroup",file=file)
        print(node.rna,file=file)
    
        
````

````python


#mutation rate in hours
mutrate = 1/200
mutcount=0
pop_today=1
seqlen=5000
datestart=datetime(2020,2,1)
growthrate=1.2
totdays=60



# Generate Starting sequence
root=RNA(index=-1, pop=0, nuc=0, time=0, country= "france")

for i in range(0,seqlen):
    root.rnaint[i]= random.randrange(4)
root.rna_string()

#First infection
initRNA = RNA(index=0, pop=1, nuc=0, time=-60, country= "france")
root.add_child(initRNA)
genomelist = [initRNA]
initRNA.populate(totdays,0)

#Simulate Loop
for d in range(0,totdays):
    #print(d, end="")
    mutates= random.random()
  
    if mutates <(pop_today)*mutrate:
        mutcount=mutcount+1
        prob=[]
        for rna in genomelist:
            prob.append(rna.Pop[d]/pop_today)
                    
        whichparent=random.choices(genomelist, prob)[0]
        newsequence=RNA(index=mutcount, pop=1, nuc=random.randrange(0,seqlen), time=d, country= "france")
        newsequence.populate(totdays,d)
        genomelist.append(newsequence)
        whichparent.add_child(newsequence)
    pop_today=0   
    for rna in genomelist:
        pop_today= pop_today + rna.Pop[d+1]
        
             


````

````python
print("mutations: ",mutcount, "\ntotal infected:", int(pop_today))
mutate_tree(initRNA)
print("hi")
pprint_tree(initRNA)
pprint_rna_tree(initRNA)
fasta = open('data//sequences.fasta', 'w') 
nextstrain_print(initRNA,file=fasta)
fasta.close()

meta= open('data//metadata.tsv','w')
print("strain"+"\t"+"virus"+"\t"+"date"+"\t"+"country",file=meta)
nextstrain_print_dat(initRNA,file=meta)
meta.close()
````
